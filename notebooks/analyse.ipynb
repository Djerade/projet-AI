```python
import nibabel as nib

# Charger le fichier NIfTI
nifti_file = nib.load("/home/perfect/Documents/GitHub/projet-AI/data/51_OtherNeoplasms/BraTS-SSA-00009-000/BraTS-SSA-00009-000-seg.nii.gz")

# Extraire les données sous forme de tableau NumPy
volume = nifti_file.get_fdata()

# Afficher la forme (dimensions)
print("Dimensions du volume :", volume.shape)
```

```python
import matplotlib.pyplot as plt


slice_index = volume.shape[2] // 2

# Afficher la coupe
plt.imshow(volume[:, :, slice_index], cmap="gray")
plt.title(f"Coupe axiale à l'indice {slice_index}")
plt.axis("off")
plt.show()
```
